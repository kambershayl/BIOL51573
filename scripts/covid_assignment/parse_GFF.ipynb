```python
!/usr/bin/env python3

import argparse

def parse_args():
    parser = argparse.ArgumentParser(description = "this script will collect data on the name of the files")
    parser.add_argument("fasta_name", "fasta", help = "the name of the fasta file you want to analyze")
    parser.add_argument("gff_name", "gff", help = "the name of the gff file you want to analyze")
    args = parser.parse_args()
    return args

def main():
    args = parse_args()
    print(args.fasta_name)
    print(args.gff_name)


if __name__ == '__main__':
    main()
    
```
