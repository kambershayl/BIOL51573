```python
# remind ourselves of the list
print(beatles)
# print the first beatle (John)
print(beatles[0])
# many ways to print the last Beatle
print(beatles[-1])
print(beatles[4])
print(beatles(len(beatles)-1))
# many ways to print George
print(beatles[2])
print(beatles[-3])
print(beatles(len(beatles)-3))
```

### Make some new lists

```python
# start fresh with new groups
jackson5 = ["jackie", "tito", "jermaine","marlon", "michael"]
destinyschild = ["beyonce", "kelly", "michell"]

# first remove Michael :(
del(jackson5[-1])
```

### Growing a list with `extend()`

```python
# print our lists
print(jackson5)
print(destinyschild)
# this is doing the equivalent of jackson5 + destinyschild, but this would create a new list
# use extend() to add Destiny's Child to the Jackson 5
jackson5.extend(destinyschild)
print(f"here's the extended jackson 5 {jackson5}")
```

>so `extend()` modifies a list in place, it doesn't return a value (i.e., a new list)

>if you want a new list, add them together (`new_list = list1 + list2`) instead of extending

### Growing a list with `append()`

```python
# use `append()` to add lists a different way
num = [1, 2, 3]
let = ("a", "b", "c")

print(f"our lists: {let, num}")

# append let onto num
num.append(let)

print(f"after appending let to num: {num}")

# used mixed data types, made the appended list as a sublist. 
# append will append the item as a single item each time

more_num = [4, 5, 6]
more_num.append(7)
i = 8
more_num.append(i)
print(more_num)
```

>What did append do?

>It adds a single item, even if that item is a list.
